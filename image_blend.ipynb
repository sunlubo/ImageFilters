```python
from PIL import Image
import numpy as np
import math
```

```python
base = Image.open("img/jiuzhaigou.jpg")
overlay = Image.open("img/girl.jpg")
```

```python
def LightenBlend(base, overlay):
    """
    Blends two images by taking the maximum value of each color component between the images.
    """
    output = Image.new(mode='RGB', size=base.size)
    for i in range(base.height):
        for j in range(base.width):
            p1 = base.getpixel((j, i))
            p2 = overlay.getpixel((j, i))
            output.putpixel((j, i), tuple(map(max, p1, p2)))
    return output

LightenBlend(base, overlay)
```

```python
def DarkenBlend(base, overlay):
    output = Image.new(mode='RGB', size=base.size)
    for i in range(base.height):
        for j in range(base.width):
            p1 = base.getpixel((j, i))
            p2 = overlay.getpixel((j, i))
            output.putpixel((j, i), tuple(map(min, p1, p2)))
    return output

DarkenBlend(base, overlay)
```

```python
def DissolveBlend(base, overlay, mix=0.5):
    """
    Applies a dissolve blend of two images.

    :param mix: Mix ranges from 0.0 (only image 1) to 1.0 (only image 2), with 0.5 (half of either) as the normal level
    """
    output = Image.new(mode='RGB', size=base.size)
    for i in range(base.height):
        for j in range(base.width):
            p1 = np.array(base.getpixel((j, i)))
            p2 = np.array(overlay.getpixel((j, i)))
            output.putpixel((j, i), tuple(map(int, p1 + (p2 - p1) * mix)))
    return output

DissolveBlend(base, overlay)
```

```python
def DifferenceBlend(base, overlay):
    """
    Applies a difference blend of two images.
    """
    output = Image.new(mode='RGB', size=base.size)
    for i in range(base.height):
        for j in range(base.width):
            p1 = np.array(base.getpixel((j, i)))
            p2 = np.array(overlay.getpixel((j, i)))
            output.putpixel((j, i), tuple(map(int, abs(p2 - p1))))
    return output

DifferenceBlend(base, overlay)
```
