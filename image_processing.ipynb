```python
def Crop(im, region):
    if region is None:
        return im
    
    assert region[0] + region[2] <= im.width
    assert region[1] + region[3] <= im.height

    output = Image.new(mode='RGB', size=(region[2], region[3]))
    for i in range(region[1], region[1] + region[3]):
        for j in range(region[0], region[0] + region[2]):
            output.putpixel((j - region[0], i - region[1]), im.getpixel((j, i)))
    return output

Crop(im, region=(int((im.width - 200) / 2), int((im.height - 200) / 2), 200, 200))
```

```python
def Convolution3x3(im, matrix=[0, 0, 0, 0, 1, 0, 0, 0, 0]):
    output = Image.new(mode='RGB', size=im.size)
    width, height = im.size
    for i in range(height):
        for j in range(width):
            bottom = (0, 0, 0) if i + 1 == height else im.getpixel((j, i + 1))
            bottomLeft = (0, 0, 0) if i + 1 == height or j - 1 < 0 else im.getpixel((j - 1, i + 1))
            bottomRight = (0, 0, 0) if i + 1 == height or j + 1 == width else im.getpixel((j + 1, i + 1))
            center = im.getpixel((j, i))
            left = (0, 0, 0) if j - 1 < 0 else im.getpixel((j - 1, i))
            right = (0, 0, 0) if j + 1 == width else im.getpixel((j + 1, i))
            top = (0, 0, 0) if i - 1 < 0 else im.getpixel((j, i - 1))
            topRight = (0, 0, 0) if j + 1 == width or i - 1 < 0 else im.getpixel((j + 1, i - 1))
            topLeft = (0, 0, 0) if j - 1 < 0 or i - 1 < 0 else im.getpixel((j - 1, i - 1))

            r = topLeft[0] * matrix[0] + top[0] * matrix[1] + topRight[0] * matrix[2] 
            r += left[0] * matrix[3] + center[0] * matrix[4] + right[0] * matrix[5]
            r += bottomLeft[0] * matrix[6] + bottom[0] * matrix[7] + bottomRight[0] * matrix[8]

            g = topLeft[1] * matrix[0] + top[1] * matrix[1] + topRight[1] * matrix[2] 
            g += left[1] * matrix[3] + center[1] * matrix[4] + right[1] * matrix[5]
            g += bottomLeft[1] * matrix[6] + bottom[1] * matrix[7] + bottomRight[1] * matrix[8]

            b = topLeft[2] * matrix[0] + top[2] * matrix[1] + topRight[2] * matrix[2] 
            b += left[2] * matrix[3] + center[2] * matrix[4] + right[2] * matrix[5]
            b += bottomLeft[2] * matrix[6] + bottom[2] * matrix[7] + bottomRight[2] * matrix[8]  

            f = lambda x: int(min(max(x, 0), 255))
            output.putpixel((j, i), tuple(map(f, [r, g, b])))       

    return output

Convolution3x3(im, matrix=[-1, 0, 1, -2, 0, 2, -1, 0, 1])
```
